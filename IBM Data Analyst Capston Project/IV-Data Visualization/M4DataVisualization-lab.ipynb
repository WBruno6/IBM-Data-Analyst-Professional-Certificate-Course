# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# your code goes here
import matplotlib as mpl
import matplotlib.pyplot as plt

QUERY = """
SELECT ConvertedComp
FROM master
"""
ccmp = pd.read_sql_query(QUERY, conn)
ccmp.plot(kind='hist')
plt.title("Histogram of ConvertedComp")
plt.ylabel("Count")
plt.xlabel("ConvertedComp")
plt.show()
```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here
QUERY = """
SELECT Age
FROM master
"""
age = pd.read_sql_query(QUERY, conn)
age.plot(kind='box')
plt.title("Box plot of Age")
plt.ylabel("Age Distribution")
plt.show()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here
QUERY = """
SELECT Age, WorkWeekHrs
FROM master
"""
df_aw = pd.read_sql_query(QUERY, conn)
df_aw.plot(kind='scatter', x='Age', y='WorkWeekHrs')
plt.title("scatter plot of Age and WorkWeekHrs")
plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here
import seaborn as sns
QUERY = """
SELECT Age, WorkWeekHrs, CodeRevHrs
FROM master
"""
df_awc = pd.read_sql_query(QUERY, conn)
plt.figure(figsize=(12, 8))
sns.scatterplot(x="WorkWeekHrs", y="CodeRevHrs", size="Age", sizes=(20,300), hue="Age", data=df_awc);
plt.title("scatter plot of Age and WorkWeekHrs")
plt.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# your code goes here
QUERY = """
SELECT DatabaseDesireNextYear, COUNT(*) AS count_DB
FROM DatabaseDesireNextYear 
GROUP BY DatabaseDesireNextYear
ORDER BY count_DB DESC
LIMIT 5

"""

db_top5 = pd.read_sql_query(QUERY,conn)
explode_list = [0.04, 0.03, 0.02, 0.01, 0]
plt.figure(figsize=(8,8))
plt.pie(db_top5["count_DB"],
        labels=db_top5["DatabaseDesireNextYear"].values,
        autopct="%1.1f%%",
        startangle=90,    
        shadow=True,
        explode=explode_list)
plt.title("Top 5 Databases that Respondents desire to Learn Next Year",fontsize= 10)
plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here
QUERY = """
SELECT WorkWeekHrs, CodeRevHrs, Age
FROM master
WHERE AGE BETWEEN 30 AND 35
"""
data_age_group = pd.read_sql_query(QUERY,conn)

median_work_week_hours = data_age_group.groupby('Age')['WorkWeekHrs'].median()
median_code_review_hours = data_age_group.groupby('Age')['CodeRevHrs'].median()
plt.figure(figsize=(10,8))
plt.bar(median_work_week_hours.index, median_work_week_hours, label='WorkWeekHrs')
plt.bar(median_code_review_hours.index, median_code_review_hours, bottom=median_work_week_hours, label='CodeRevHrs')

plt.xlabel('Age Group')
plt.ylabel('Median Hours')
plt.title('Median WorkWeekHrs and CodeRevHrs for Age Group 30-35')
plt.legend(loc = (1, 0.93))
plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here
QUERY = """
SELECT ConvertedComp, Age
FROM master
WHERE AGE BETWEEN 40 AND 60
"""
data_age_con = pd.read_sql_query(QUERY,conn)

median_converted_comp = data_age_con.groupby('Age')['ConvertedComp'].median()

plt.figure(figsize=(10,8))
plt.plot(median_converted_comp.index, median_converted_comp, marker='s', color='green')
plt.xlabel('Age')
plt.ylabel('Median Converted Compensation')
plt.title('Median Converted Compensation for Ages 45-60')
plt.grid(False)
plt.show()

```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here
QUERY = """
SELECT MainBranch, COUNT(*) AS Total
FROM master
GROUP BY MainBranch
"""
main_branch_counts = pd.read_sql_query(QUERY,conn)

plt.figure(figsize=(10,8))
sns.barplot(y=main_branch_counts['MainBranch'], x=main_branch_counts['Total'], orient='h')

plt.xlabel('Count')
plt.ylabel('Main Branch')
plt.title('Distribution of Main Branches')

plt.show()
```
